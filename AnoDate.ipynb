```python
import datetime, time
import random
from random import randrange
def rand_id(long):
    if(long=="DEL"):
        return long
    else:
        return long[11:20]
    
df['id']=df['id'].apply(lambda x: rand_id(x))
```

```python
import datetime, time
import random
from random import randrange
def rand_lat(lat):
    r=randrange(2)
    if(r==0):
        return lat + random.uniform(0, 0.009)
    else:
        return lat - random.uniform(0, 0.009)
    



df['lat']=df['lat'].apply(lambda x: rand_lat(x))
```

```python
import datetime, time
import random
from random import randrange
def rand_long(long):
    r=randrange(2)
    if(r==0):
        return long + random.uniform(0, 0.009)
    else:
        return long - random.uniform(0, 0.009)
    



df['long']=df['long'].apply(lambda x: rand_long(x))
```
