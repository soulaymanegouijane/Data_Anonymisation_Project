```python
import pandas as pd
df = pd.read_csv("ano_pos_minsec_occ.csv",delimiter='\t',names=['id','datetime','lat','long',"week","ordre"],low_memory=False)
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
import datetime, time

def set_seconds_module(date):
    match= date.hour
    if (match == 7):
        return date.replace(hour = 6, minute =0,second=0 )
    elif (match == 8):
        return date.replace(hour = 9, minute =0,second=0 )
    elif (match == 17):
        return date.replace(hour = 16, minute =0,second=0 )
    elif (match == 18):
        return date.replace(hour = 16, minute =0,second=0 )
    elif (match == 19):
        return date.replace(hour = 22, minute =0,second=0 )
    elif (match == 20):
        return date.replace(hour = 22, minute =0,second=0 )
    elif (match == 21):
        return date.replace(hour = 22, minute =0,second=0 )
    else:
        return date.replace(minute =0,second=0 )



df['datetime']=df['datetime'].apply(lambda x: set_seconds_module(x))
```
